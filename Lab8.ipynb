```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
df=pd.read_csv("all_alpha_19.csv", header="infer")
df.head(10)
```

```python
q1=df[(df.Stnd =="T3B125") & ((df.Fuel=="Diesel") |(df.Fuel=="Gasoline"))]
q1.head(10)
```

```python
cols=["Model", "Displ","Fuel","City MPG","Hwy MPG","Cmb MPG","Greenhouse Gas Score"]
new_df = q1[cols]
new_df.reset_index(drop=True).head(10)
```

```python
new_df.dtypes
```

```python
new_df[["City MPG", "Hwy MPG", "Cmb MPG"]]= new_df[["City MPG", "Hwy MPG", "Cmb MPG"]].astype(float)
```

```python
new_df.dtypes
```

```python
df2=new_df.assign(CityKML=lambda x: x["City MPG"] *0.42514,
           HwyKML=lambda x: x["Hwy MPG"] *0.42514,
           CmbKML=lambda x: x["Cmb MPG"] *0.42514)
df2.head(10)
```

```python
df2.to_csv("car_data.csv", header=True, index=True)
```

```python
df = pd.read_csv("car_data.csv", header="infer",index_col=0)
```

```python
df.plot(kind="scatter",x="Displ", y="City MPG", color="r")
plt.show()
```

```python
df["c"] = ["r" if x == "Gasoline" else "g" for x in df["Fuel"]]
df["s"]=df["Greenhouse Gas Score"]*8
```

```python
df.plot(kind="scatter",x="Displ", y="City MPG",color=df["c"], s=df["s"],alpha=0.5)
plt.xlabel("Engine displacement ")
plt.ylabel("Number of miles per gallon of fuel in the city")
plt.title("Fuel efficiency")
red_patch = mpatches.Patch(color="r", label="Gasoline")
green_patch = mpatches.Patch(color="g", label="Diesel")
plt.legend(handles=[red_patch, green_patch])
plt.show()
```

```python
ax = plt.gca()# gca stands for 'get current axis'
df.plot(kind="line",x="Displ",y="Hwy MPG",color="r", label= "Hwy MPG",ax=ax)
df.plot(kind="line",x="Displ",y="Cmb MPG",color="y", label= "Cmb MPG",ax=ax)
df.plot(kind="line",x="Displ",y="City MPG",color="b", label= "City MPG",ax=ax)
df.plot(kind="line",x="Displ",y="HwyKML",color="g", label= "HwyKML",ax=ax)
df.plot(kind="line",x="Displ",y="CmbKML",color="c", label= "CmbKML",ax=ax)
df.plot(kind="line",x="Displ",y="CityKML",color="m", label= "CityKML",ax=ax)
plt.xlabel("Engine displacemnt")
plt.ylabel("City, highway and combined MPG/KML")
plt.title("Fuel efficiency")
plt.legend()
plt.show()
```
